2.5 between time
Create a time series of events encompassing an entire day, separated by 10 minutes, and use the between time function to select events that occurred between
09h40 and 12h20. Print the results (head only, 10 entries).


```python
import pandas as pd
from datetime import datetime, timedelta

start_time = datetime(2023, 1, 1, 0, 0) 
end_time = datetime(2023, 1, 2, 0, 0)
time_series = pd.date_range(start=start_time, end=end_time, freq='10T')

events_df = pd.DataFrame(index=time_series)
selected_events = events_df.between_time('09:40', '12:20')
selected_events
```
